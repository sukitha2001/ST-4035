```python
# Remove rows where student_id is missing
survey = survey.dropna(subset=['student_id'])

# Impute numeric missing with the median
survey['Score'] = survey['Score'].fillna(survey['Score'].median())

# Impute character missing with "Unknown"
survey['Name'] = survey['Name'].fillna('Unknown')
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv'
df = pd.read_csv(url)
print(df.head())
```

```python
df_07 = df[df['year'] == 2007]
df_07.head()
```

```python
(
    df_07
    .groupby(['continent'])
    .agg(
        avg_lifeExp = ('lifeExp',   'mean'),
        avg_gdp     = ('gdpPercap', 'mean')
    )
    .reset_index()
)
```

Oceania has the higest average gpd while Africa has the lowest average life expectency

```python
df['gdp_billion'] = (df['gdpPercap'] * df['pop']) / 1e9

df['decade'] = (df['year'] // 10 * 10).astype(str) + 's'

top_5_2007 = df[df['year'] == 2007].nlargest(5, 'gdp_billion')

print(top_5_2007[['country', 'gdp_billion']])
```
