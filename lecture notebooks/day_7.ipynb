### 1.VIF stands for Variance Inflation Factor
### 2. Interpretation
VIF is the ratio of the variance of a coefficient in the full model to the variance of that same coefficient if it were modeling the target alone.
VIF = 1: No correlation.
VIF between 1 and 5: Moderate correlation, usually not severe.
VIF > 5 or 10: High multicollinearity, suggesting the regression coefficients are poorly estimated.
### 3. Formula
VIF = 1/(1-R^2)
