```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from itertools import combinations
```

```python
# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names
```

```python
# Get all unique pairs of classes (0,1), (0,2), (1,2)
class_pairs = list(combinations(np.unique(y), 2))

for class_a, class_b in class_pairs:
    # Filter dataset for the two specific classes
    mask = (y == class_a) | (y == class_b)
    X_pair = X[mask]
    y_pair = y[mask]
    
    # Split into 80% train and 20% test
    X_train, X_test, y_train, y_test = train_test_split(
        X_pair, y_pair, test_size=0.20, random_state=42
    )
    # Initialize and fit SVM model
    # Using linear kernel for binary classification
    model = SVC(kernel='linear')
    model.fit(X_train, y_train)
    
    # Predictions
    y_pred = model.predict(X_test)
    
    # Output results
    print(f"\n{'='*60}")
    print(f"SVM Classification: {target_names[class_a]} vs {target_names[class_b]}")
    print(f"{'='*60}")
    print(classification_report(y_test, y_pred, target_names=[target_names[class_a], target_names[class_b]]))
```
